```python
import numpy as np
from sklearn import linear_model, datasets
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt


class fit_iris_dataset:


    def __init__(self):
        self.iris = datasets.load_iris()
        self.X    = self.iris.data
        self.Y    = self.iris.target

    def show_x_y(self):
        print("Printing first 10 samples of X and Y")
        print(self.X[0:10,:])
        print("--------------------")
        print("Printing first 10 features")
        print(self.Y[0:10])

    def fit_model(self, grid_number):
        grid_C = np.linspace(0.01, 1, grid_number)
        parameters = {'C': grid_C}
        print("Parameters to be tested")
        print(parameters)
        logreg = linear_model.LogisticRegression(C=1e5, penalty='l2', class_weight=None, fit_intercept = True)
        clf = GridSearchCV(logreg, parameters)
        clf.fit(self.X, self.Y)
        print(clf.best_score_)
        print(clf.best_params_)
        self.best_model = clf.best_estimator_

        M1 = np.concatenate([clf.cv_results_["split0_test_score"].reshape(-1, 1), clf.cv_results_["split1_test_score"].reshape(-1, 1), clf.cv_results_["split2_test_score"].reshape(-1, 1)], axis=1)
        print("cross validation results for each C parameter on the grid")
        print(M1)

        mean = np.mean(M1, axis=1)

        print("mean cross validation results for each C parameter on the grid")
        print(mean)

        fig, ax = plt.subplots()
        ax.plot(grid_C, mean)

        ax.set(xlabel='C', ylabel='Mean cross validation score', title='Mean cross validation score vs C')
        ax.grid()
        plt.show()
```

```python
model_ = fit_iris_dataset()
```

```python
model_.show_x_y()
```

```python
model_.fit_model(10)
```

```python
model_.fit_model(100)
```
